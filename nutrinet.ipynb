```python
# Dane Brakujące: Zidentyfikuj kolumny, w których brakuje więcej niż 50% wartości i rozważ ich usunięcie, jeśli są nieistotn

lista = (df.isna().sum()/df['fdc_id'].count())*100
lista[lista>50]
```

```python
# Typy Danych: Sprawdź typy danych w każdej kolumnie. Konwertuj kolumny na odpowiednie typy danych, jeśli to konieczne, na przykład modified_date i available_date na typ daty.

df['modified_date'] = pd.to_datetime(df['modified_date'])
df['available_date'] = pd.to_datetime(df['available_date'])
```

```python
#Unikalne Wartości: Znajdź liczbę unikalnych wartości dla brand_owner.
df['brand_owner'].unique()
```

```python
# Zliczanie wystąpień każdej kategorii w kolumnie 'branded_food_category'
category_counts = df['branded_food_category'].value_counts()
most_common_category = category_counts.idxmax()
most_common_count = category_counts.max()
print(f"Najczęściej występująca kategoria to '{most_common_category}' z liczbą wpisów: {most_common_count}")
```
