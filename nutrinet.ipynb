```python
#Unikalne WartoÅ›ci: ZnajdÅº liczbÄ™ unikalnych wartoÅ›ci dla brand_owner.
df['brand_owner'].unique()
```

```python
# Zliczanie wystÄ…pieÅ„ kaÅ¼dej kategorii w kolumnie 'branded_food_category'
category_counts = df['branded_food_category'].value_counts()
most_common_category = category_counts.idxmax()
most_common_count = category_counts.max()
print(f"NajczÄ™Å›ciej wystÄ™pujÄ…ca kategoria to '{most_common_category}' z liczbÄ… wpisÃ³w: {most_common_count}")
```

```python
# Podsumowanie Statystyczne: Podaj podsumowanie statystyczne (Å›rednia, mediana, moda, odchylenie standardowe) dla kolumn Total lipid (fat)-G i Vitamin C, total ascorbic acid-MG

df[['Total lipid (fat)-G', 'Vitamin C, total ascorbic acid-MG']].describe()
```

```python
# Pokarmy Bogate w TÅ‚uszcz: Wyfiltruj dane pokazujÄ…ce tylko te wpisy, gdzie Total lipid (fat)-G przekracza 20 gramÃ³w na porcjÄ™

df.loc[df['Total lipid (fat)-G'] > 20]
```

```python
# Nowe Produkty: PokaÅ¼ tylko produkty zmodyfikowane w ostatnim kwartale 2020 roku.
df.loc[(df['modified_date'] >= '2020-09-01') & (df['modified_date'] <= '2020-12-31')]
```

```python
# Åšrednie SkÅ‚adniki OdÅ¼ywcze wedÅ‚ug Kategorii: Oblicz Å›redniÄ… iloÅ›Ä‡ Vitamin D (D2 + D3), International Units-IU i Total lipid (fat)-G grupujÄ…c wedÅ‚ug branded_food_category.

df.groupby('branded_food_category')[['Vitamin D (D2 + D3), International Units-IU','Total lipid (fat)-G']].mean()
```

```python
# Suma Witaminy A wedÅ‚ug Marki: Oblicz sumÄ™ Vitamin A, IU-IU dla kaÅ¼dego brand_owner.

df.groupby('brand_owner')['Vitamin A, IU-IU'].sum()
```

```python
# Tabela Przestawna: StwÃ³rz tabelÄ™ przestawnÄ… pokazujÄ…cÄ… Å›redniÄ… Vitamin C, total ascorbic acid-MG dla kaÅ¼dego brand_owner, z indeksowaniem wedÅ‚ug branded_food_category

table = pd.pivot_table(df, values='Vitamin C, total ascorbic acid-MG', index='branded_food_category', aggfunc='mean')
table
```

```python
# Sortowanie Wielopoziomowe: Posortuj dane najpierw wedÅ‚ug serving_size malejÄ…co, a nastÄ™pnie Total lipid (fat)-G rosnÄ…co w ramach kaÅ¼dej serving_size.

df.sort_values(by=['serving_size','Total lipid (fat)-G'], ascending=[False, True])
```

```python
# Imputacja BrakujÄ…cych WartoÅ›ci: Dla kolumn Vitamin B-12-UG i Zinc, Zn-MG, zastÄ…p braki Å›redniÄ… z ich kolumn.
Vitamin_B12_UG = df['Vitamin B-12-UG'].mean()
Zinc_ZnMG = df['Zinc, Zn-MG'].mean()

df['Vitamin B-12-UG'].fillna(Vitamin_B12_UG, inplace=True)
df['Zinc, Zn-MG'].fillna(Zinc_ZnMG, inplace=True)
```
