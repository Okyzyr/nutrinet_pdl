```python
import pandas as pd
```

```python
df = pd.read_csv('nutrinet.csv')
```

```python
df.sample(4)
```

```python
# Dane Brakujące: Zidentyfikuj kolumny, w których brakuje więcej niż 50% wartości i rozważ ich usunięcie, jeśli są nieistotn

lista = (df.isna().sum()/df['fdc_id'].count())*100
lista[lista>50]
```

```python
# Typy Danych: Sprawdź typy danych w każdej kolumnie. Konwertuj kolumny na odpowiednie typy danych, jeśli to konieczne, na przykład modified_date i available_date na typ daty.

df['modified_date'] = pd.to_datetime(df['modified_date'])
df['available_date'] = pd.to_datetime(df['available_date'])
```

```python
#Unikalne Wartości: Znajdź liczbę unikalnych wartości dla brand_owner.
df['brand_owner'].unique()
```

```python
# Zliczanie wystąpień każdej kategorii w kolumnie 'branded_food_category'
category_counts = df['branded_food_category'].value_counts()
most_common_category = category_counts.idxmax()
most_common_count = category_counts.max()
print(f"Najczęściej występująca kategoria to '{most_common_category}' z liczbą wpisów: {most_common_count}")
```

```python
df.columns
```

```python
# Podsumowanie Statystyczne: Podaj podsumowanie statystyczne (średnia, mediana, moda, odchylenie standardowe) dla kolumn Total lipid (fat)-G i Vitamin C, total ascorbic acid-MG

df[['Total lipid (fat)-G', 'Vitamin C, total ascorbic acid-MG']].describe()
```

```python
# Pokarmy Bogate w Tłuszcz: Wyfiltruj dane pokazujące tylko te wpisy, gdzie Total lipid (fat)-G przekracza 20 gramów na porcję

df.loc[df['Total lipid (fat)-G'] > 20]
```

```python
# Nowe Produkty: Pokaż tylko produkty zmodyfikowane w ostatnim kwartale 2020 roku.
df.loc[(df['modified_date'] >= '2020-09-01') & (df['modified_date'] <= '2020-12-31')]
```

```python
# Sortowanie po Dacie: Posortuj dane według modified_date malejąco.
df.sort_values(by='modified_date', ascending=False)
```

```python
df.columns
```

```python
# Średnie Składniki Odżywcze według Kategorii: Oblicz średnią ilość Vitamin D (D2 + D3), International Units-IU i Total lipid (fat)-G grupując według branded_food_category.

df.groupby('branded_food_category')[['Vitamin D (D2 + D3), International Units-IU','Total lipid (fat)-G']].mean()
```

```python
# Suma Witaminy A według Marki: Oblicz sumę Vitamin A, IU-IU dla każdego brand_owner.

df.groupby('brand_owner')['Vitamin A, IU-IU'].sum()
```

```python
# Tabela Przestawna: Stwórz tabelę przestawną pokazującą średnią Vitamin C, total ascorbic acid-MG dla każdego brand_owner, z indeksowaniem według branded_food_category

table = pd.pivot_table(df, values='Vitamin C, total ascorbic acid-MG', index='branded_food_category', aggfunc='mean')
table
```

```python
# Sortowanie Wielopoziomowe: Posortuj dane najpierw według serving_size malejąco, a następnie Total lipid (fat)-G rosnąco w ramach każdej serving_size.

df.sort_values(by=['serving_size','Total lipid (fat)-G'], ascending=[False, True])
```

```python
# Imputacja Brakujących Wartości: Dla kolumn Vitamin B-12-UG i Zinc, Zn-MG, zastąp braki średnią z ich kolumn.
Vitamin_B12_UG = df['Vitamin B-12-UG'].mean()
Zinc_ZnMG = df['Zinc, Zn-MG'].mean()

df['Vitamin B-12-UG'].fillna(Vitamin_B12_UG, inplace=True)
df['Zinc, Zn-MG'].fillna(Zinc_ZnMG, inplace=True)
```
